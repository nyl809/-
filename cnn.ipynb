```python
import os
import cv2
import numpy as np
import keras
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import time
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt


# Path setup
output_dir = '/root/autodl-tmp/lfw'
threshold = 100  # Dataset threshold for picking
images, labels, names = [], [], []
label = 1

# Load images and labels
for name in os.listdir(output_dir):
    class_dir = os.path.join(output_dir, name)
    if len(os.listdir(class_dir)) > threshold:
        for file in os.listdir(class_dir):
            file_path = os.path.join(class_dir, file)
            img = cv2.imread(file_path)
            img = cv2.resize(img, (200, 200))  # Resize to 200x200
            images.append(img)
            labels.append(label)
        names.append(name)
        label += 1

images = np.array(images)
labels = np.array(labels)

# Train-test split
images_train, images_test, labels_train, labels_test = train_test_split(images[:1000], labels[:1000], test_size=0.1)
class_number = len(names) + 1
images_train = images_train.astype('float32') / 255.0
images_test = images_test.astype('float32') / 255.0

labels_train_b = keras.utils.to_categorical(labels_train, class_number)
labels_test_b = keras.utils.to_categorical(labels_test, class_number)

def plot_confusion_matrix(y_true, y_pred_class, class_names):
    cm = confusion_matrix(y_true, y_pred_class)
    
    # 可视化混淆矩阵
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
    
    return cm

# Common training and testing function
def train_and_test_model(optimizer, model_name, images_train, labels_train_b, images_test, labels_test_b):
    # Train
    start_time = time.time()
    model = CNNModel(batchsize=32, epochs=100, classnumber=class_number, optimizer=optimizer)
    model.build_model()
    model.train_model(images_train, labels_train_b)
    train_time = time.time() - start_time
    print(f"{model_name} train_time: {train_time} seconds")
    model.save_model(f'{model_name}_model')
    model.save_para(f'TrainAcc_{model_name}', f'TrainLoss_{model_name}')

    # Test
    start_time = time.time()
    loss, accuracy = model.evaluate(images_test, labels_test_b)
    test_time = time.time() - start_time
    print(f"{model_name} test_time: {test_time} seconds")
    np.savetxt(f'Test_{model_name}', [loss, accuracy])

    # 获取预测结果
    y_pred = model.model.predict(images_test)  # 获取每个类的预测概率
    y_pred_class = np.argmax(y_pred, axis=1)  # 获取预测的类别
    y_true = np.argmax(labels_test_b, axis=1)  # 获取真实标签

    # 绘制混淆矩阵
    class_names = [str(i) for i in range(class_number)]  # 类别名称
    plot_confusion_matrix(y_true, y_pred_class, class_names)

    # Visualize the training process
    model.visualize_training_process(model_name)

# Model setup
class CNNModel:
    def __init__(self, batchsize, epochs, classnumber, optimizer):
        self.model = None
        self.batchsize = batchsize
        self.epochs = epochs
        self.classnumber = classnumber
        self.optimizer = optimizer
        self.history = None  # Store history for visualization

    def build_model(self):
        self.model = Sequential([
            Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(200, 200, 3)),
            MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'),
            Dropout(0.25),
            Conv2D(64, (3, 3), padding='same', activation='relu'),
            MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='same'),
            Dropout(0.25),
            Flatten(),
            Dense(512, activation='relu'),
            Dense(self.classnumber, activation='softmax')
        ])
        self.model.summary()

    def train_model(self, images, labels):
        self.model.compile(optimizer=self.optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
        self.history = self.model.fit(images, labels, epochs=self.epochs, batch_size=self.batchsize)

    def evaluate(self, images_test, labels_test):
        print('\nTesting---------------')
        loss, accuracy = self.model.evaluate(images_test, labels_test)
        print(f'test loss: {loss}, test accuracy: {accuracy}')
        return loss, accuracy

    def save_model(self, filename):
        self.model.save(filename)
        print('Model saved')

    def save_para(self, accuracy_name, loss_name):
        np.savetxt(accuracy_name, np.array(self.history.history['accuracy']))
        np.savetxt(loss_name, np.array(self.history.history['loss']))

    def visualize_training_process(self, model_name):
        # Plot the training accuracy and loss
        plt.figure(figsize=(12, 5))

        # Plot training accuracy
        plt.subplot(1, 2, 1)
        plt.plot(self.history.history['accuracy'], label='Train Accuracy')
        plt.title(f'Train Accuracy ({model_name})')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        # Plot training loss
        plt.subplot(1, 2, 2)
        plt.plot(self.history.history['loss'], label='Train Loss')
        plt.title(f'Train Loss ({model_name})')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        # Display the plots
        plt.tight_layout()
        plt.show()

# Training and testing using different optimizers
train_and_test_model(keras.optimizers.RMSprop(learning_rate=0.0001, decay=1e-6), 'RMSprop', images_train, labels_train_b, images_test, labels_test_b)
train_and_test_model(keras.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True), 'SGD', images_train, labels_train_b, images_test, labels_test_b)
train_and_test_model(keras.optimizers.Adam(lr=0.001), 'Adam', images_train, labels_train_b, images_test, labels_test_b)
```
